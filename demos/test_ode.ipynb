```python
implicit = pd.read_csv(f"{path}\\implicit_mass_spring.csv")
explicit = pd.read_csv(f"{path}\\explicit_mass_spring.csv")
crank_nicolson = pd.read_csv(f"{path}\\crank_nicolson_mass_spring.csv")

plt.figure()
plt.plot(implicit["t"], implicit["y[0]"], label="implicit")
plt.plot(explicit["t"], explicit["y[0]"], label="explicit")
plt.plot(crank_nicolson["t"], crank_nicolson["y[0]"], label="crank nicolson")
plt.grid()
plt.legend()
plt.show()
```
